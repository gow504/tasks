```python
#cabin column is drop becuase tooo much missing values
data.drop(columns='Cabin',inplace=True)
```

```python
 #2.Create a numeric features.

 numeric_features = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

```

```python
numeric_features
```

```python
corr_matrix = data[numeric_features].corr()

```

```python
corr_matrix
```

```python
target_corr = corr_matrix["Survived"].sort_values(ascending=False)
print("Correlation of features with Survived:\n")
print(target_corr)

```

```python
for feature in numeric_features:
    plt.figure(figsize=(4, 3))
    sns.boxplot(data=data, x=feature)
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

```python
for feature in numeric_features:
    plt.figure(figsize=(4,3))
    sns.histplot(data=data, x=feature, kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# 3.Use pairplot/correlation matrix for feature relationships
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```

```python
 #4.Identify patterns, trends, or anomalies in the data.

#survival_by_pclass

survival_by_pclass = data.groupby('Pclass')['Survived'].mean()
print("\nSurvival rate by class:\n", survival_by_pclass)
```

```python
# Survival by sex


survival_by_sex = data.groupby('Sex')['Survived'].mean()
print("\nSurvival rate by sex:\n", survival_by_sex)

```

```python
# Survival by Embarked
survival_by_port = data.groupby('Embarked')['Survived'].mean()
print("\nSurvival rate by Embarked:\n", survival_by_port)

```

```python
 #4.Identify patterns, trends, or anomalies in the data.

#survival_by_pclass

survival_by_age = data.groupby('Age')['Survived'].mean()
print("\nSurvival rate by class:\n", survival_by_age)
```

INSIGHTS FOR TITANIC DATASET

1. Women had higher survival than men.


2. 1st class passengers had higher survival rates than 3rd class.


3. Children (young age) were more likely to survive.
4. Missing age values are anomalies that need handling.

```python
# 3. Trends
# Age distribution by Survival
plt.figure(figsize=(6,4))
sns.histplot(data=data, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Fare distribution by Survival
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Fare', data=data)
plt.title("Fare Distribution by Survival")
plt.show()

```

```python
# 5. Anomalies
# ------------------------
# Negative or unusual ages
anomalous_age = data[data['Age'] < 0]
print("\nNegative ages (anomalies):\n", anomalous_age)


```

```python
# Extremely high fares
high_fare = data[data['Fare'] > 500]
print("\nHigh fares (possible outliers):\n", high_fare)

```

```python
# Missing ages
missing_age = data[data['Age'].isnull()]
print("\nMissing Age Values:\n", missing_age)
```

```python
# 5.Make basic feature-level inferences from visuals.


plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="Survived", data=data, ci=None)
plt.title("Survival Rate by Sex")
plt.show()
```

```python
# Survival by Pclass
plt.figure(figsize=(6,4))
sns.barplot(x="Pclass", y="Survived", data=data, ci=None)
plt.title("Survival Rate by Passenger Class")
plt.show
```

```python
# Age Distribution by Survival
plt.figure(figsize=(8,5))
sns.histplot(data=data, x="Age", hue="Survived", multiple="stack", bins=30)
plt.title("Age Distribution by Survival")
plt.show
```
