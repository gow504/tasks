```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score,roc_curve,accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
data.drop(columns=['Unnamed: 32'],inplace=True)
```

```python
data.isnull().sum()
```

```python
data['diagnosis'].value_counts()
```

```python
data.replace({'diagnosis':{'M':1,'B':0}},inplace=True)
```

```python
data['diagnosis'].value_counts()
```

```python
x=data.drop(columns=['diagnosis'])
y=data['diagnosis']
```

```python
x
```

```python
print(y)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data['diagnosis'])
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(),cmap='coolwarm',cbar=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
x_pred=model.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
confusion = confusion_matrix(y_test, y_pred)
sns.heatmap(confusion, annot=True, fmt='d', cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.figure()
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc_score(y_test, y_pred):.2f})")
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
print("ROC-AUC Score:", roc_auc_score(y_test, y_pred))   # <--- ROC-AUC Score
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
